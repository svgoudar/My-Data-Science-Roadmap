### 🔍 What is Hypothesis Testing?

Hypothesis Testing is a **core concept of inferential statistics**.
It helps us draw **conclusions about a population** based on **sample data**.

 **Goal:** Use a sample to test assumptions (hypotheses) about a population parameter (like mean, proportion, etc.).

---

### 🧠 Why Do We Use It?

In many real-world cases, we **can’t measure the entire population**.
So we use sample data to make an informed judgment about the population.

---

### 📋 **Steps in Hypothesis Testing**

#### 1. **State the Null Hypothesis (H₀)**

* This is the **default assumption** (nothing unusual is happening).
* Example: "The person is **not guilty**." or "The average pass percentage is 85%."

#### 2. **State the Alternate Hypothesis (H₁ or Ha)**

* This is **what you're trying to prove** (opposite of H₀).
* Example: "The person **is guilty**." or "The average pass percentage **is not** 85%."

#### 3. **Perform Statistical Analysis**

* Collect and analyze data.
* Techniques include: **Z-test, T-test, Chi-square test, ANOVA, F-test**, etc.
* In real-world terms: like collecting **evidence** (e.g., DNA, fingerprints in a courtroom).

#### 4. **Make a Decision**

* Based on the result, either:

  * **Reject the null hypothesis**, or
  * **Fail to reject the null hypothesis** (we don’t prove H₀; we just don’t have enough evidence against it).

---

### 🧪 **Example Scenario**

**Problem**: A college in District A claims an **average pass percentage of 85%**.
A new college has a **sample of 100 students** with a **90% pass rate** and **4% standard deviation**.

**Question**: Does this new college have a different pass percentage?

#### 🧾 Hypotheses:

* **H₀ (Null)**: μ = 85 (No difference in pass percentage)
* **H₁ (Alternate)**: μ ≠ 85 (Pass percentage is different)

🔎 Since the question is asking about “a different” percentage (not specifically greater or less), it is a **two-tailed test**.

### 🔷 What is **Confidence Interval (CI)?**

A **confidence interval** gives you a **range of values** that is likely to contain the true **population parameter** (like mean or proportion) with a certain level of **confidence**.

#### ✅ Definition:

> A **confidence interval** is an estimated range of values that is likely to include an unknown population parameter, calculated from a given set of sample data.

#### 📌 Example:

Let’s say you survey 100 people and calculate the **sample mean height** as 170 cm.
You also compute a **95% confidence interval** as:

```
[168 cm, 172 cm]
```

This means:

> "We are 95% confident that the true average height of the entire population lies between 168 and 172 cm."

#### 🧠 Key Points:

* CI = Point Estimate ± Margin of Error
* Common confidence levels: **90%**, **95%**, **99%**
* A wider CI gives more certainty but less precision.
* Narrower CI gives more precision but less confidence.

---

### 🔶 What is **Significance Value (α)?**

The **significance value** (denoted by **α**) is the **threshold** used to decide whether to **reject the null hypothesis** in hypothesis testing.

#### ✅ Definition:

> The **significance value** is the maximum **probability of making a Type I error** (i.e., rejecting the null hypothesis when it's actually true) that you're willing to tolerate.

#### 💡 Common Choices:

* **α = 0.05** → 95% confidence
* **α = 0.01** → 99% confidence
* **α = 0.10** → 90% confidence

#### 🧠 Key Point:

* If your **p-value < α**, you **reject the null hypothesis**.
* If your **p-value ≥ α**, you **fail to reject the null hypothesis**.

---

### 📊 Relationship Between Confidence Interval and Significance Level

* **Confidence Level** = 1 − **Significance Level**

| Confidence Level | Significance Level (α) |
| ---------------- | ---------------------- |
| 90%              | 0.10                   |
| 95%              | 0.05                   |
| 99%              | 0.01                   |

So, a **95% confidence interval** corresponds to a **significance level α = 0.05**.

---

### 🧪 Real-Life Analogy:

Let’s say you’re testing a new medicine.

* **CI** tells you: “We are 95% confident the true improvement rate lies between 60% and 70%.”
* **α = 0.05** tells you: “We’ll only believe the medicine works if the probability of this result happening by chance is less than 5%.”



## 🎯 **Summary: Understanding p-value in Hypothesis Testing**

---

### 🔍 What is a p-value?

* A **p-value** is a number calculated from a statistical test that tells us:

  > "**How likely is it to get the observed data (or something more extreme) if the null hypothesis is true?**"

* It's a **measure of evidence** against the **null hypothesis (H₀)**.

---

### 🧠 Real-Life Analogy (Keyboard Example)

Imagine where you press the **spacebar** on a keyboard:

* Most touches are in the center → **frequent (high p-value)**
* Rare touches on the corner → **less frequent (low p-value)**

🟢 **p = 0.2** → 20 touches out of 100 → relatively common
🔴 **p = 0.01** → 1 touch out of 100 → rare event

---

### ⚖️ Hypothesis Testing Recap

1. **Null Hypothesis (H₀)**: Default assumption (e.g., coin is fair)
2. **Alternate Hypothesis (H₁)**: Opposite assumption (e.g., coin is not fair)
3. **Statistical Test** (e.g., Z-test, T-test): Produces a **p-value**

---

### 🧪 Example: Fair Coin Toss

* Tossed 100 times
* Expected: 50 Heads, 50 Tails
* Observed: 70 Heads

#### Step 1: Set Hypotheses

* **H₀**: Coin is fair
* **H₁**: Coin is not fair

#### Step 2: Define Significance Level (α)

* Typical value: **α = 0.05**
* This gives us a **95% confidence interval**
* So, only **2.5% at each tail** is considered the **rejection region**

#### Step 3: Compare p-value with α

| p-value vs α | Decision                |
| ------------ | ----------------------- |
| p ≤ α        | ❌ **Reject** H₀         |
| p > α        | ✅ **Fail to reject** H₀ |

---

### ✅ Final Takeaways

* **Low p-value** (e.g., 0.01): Strong evidence **against** H₀ → reject
* **High p-value** (e.g., 0.3): Not enough evidence → fail to reject
* P-value helps decide whether an observed result is **statistically significant**

---

## 📊 Suggested Plot: p-value and Confidence Interval Visualization

Here's a simple **plot you can show** in your video to explain p-value and rejection regions:

### **Bell Curve with Rejection Regions**

* X-axis: Number of Heads (0 to 100)
* Mean at 50
* Shaded region in the middle: **Confidence interval (e.g., 30–70)**
* Outer shaded tails: **Rejection region** (α = 0.05 → 2.5% each side)
* Example points:

  * `X = 50` → p = 0.5 (center)
  * `X = 70` → p = 0.025 (right tail)
  * `X = 90` → p < 0.001 (extreme)



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 50       # Mean (expected number of heads)
sigma = 10    # Standard deviation
alpha = 0.05  # Significance level

# X range for plotting
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Calculate the critical values for 2.5% on each tail
z_critical = norm.ppf(1 - alpha/2)
left_crit = mu - z_critical * sigma
right_crit = mu + z_critical * sigma

# Plot the normal distribution
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Normal Distribution (μ=50, σ=10)', color='blue')

# Fill the rejection regions
plt.fill_between(x, y, where=(x < left_crit), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x > right_crit), color='red', alpha=0.5, label='Rejection Region (Right)')

# Fill the acceptance region
plt.fill_between(x, y, where=(x >= left_crit) & (x <= right_crit), color='green', alpha=0.3, label='Acceptance Region (95% CI)')

# Mark an example p-value point
example_x = 70
example_p = norm.sf(example_x, mu, sigma) * 2  # Two-tailed
plt.axvline(example_x, color='black', linestyle='--', linewidth=2, label=f'Observed = 70 (p ≈ {example_p:.3f})')

# Labels and legend
plt.title("Hypothesis Testing: p-value and Rejection Region")
plt.xlabel("Number of Heads in 100 Tosses")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

### 🎯 Statistical Hypothesis Testing using **Z-Test**

---

### ✅ **Overview**:

As part of inferential statistics, we perform **hypothesis testing** to draw conclusions about a population using sample data. One such test is the **Z-test**.

---

### 📌 **Types of Statistical Tests Introduced**:

* **Z-Test**: For large sample sizes (n ≥ 30) **and** known population standard deviation.
* **T-Test**: When population standard deviation is **unknown** or sample size is **small (< 30)**.
* **Chi-Square Test**: For **categorical** data.
* **ANOVA Test**: For comparing **variances** among groups.

---

### 🧪 **When to Use a Z-Test?**

Use a **Z-Test** when:

1. Population standard deviation is **known**
2. Sample size **n ≥ 30**

---

### 📝 **Problem Statement**:

> The **population mean height** is 168 cm with a **standard deviation** of 3.9 cm.
> A doctor takes a **sample of 36 individuals** and finds their **mean height = 169.5 cm**.
> At a **95% confidence level**, determine whether the sample provides enough evidence to **reject** the claim that the population mean is 168 cm.

---

### 🔁 **Step-by-Step Hypothesis Testing**:

#### 1. **Set Hypotheses**:

* **Null Hypothesis (H₀)**: μ = 168 cm (population mean is 168 cm)
* **Alternate Hypothesis (H₁)**: μ ≠ 168 cm (mean is **not** 168 cm) → **two-tailed test**

#### 2. **Significance Level (α)**:

* CI = 95% → α = 1 − 0.95 = **0.05**
* Since it’s a two-tailed test:

  * Left tail: 0.025
  * Right tail: 0.025

#### 3. **Decision Boundary**:

* Find **critical Z-values** for α/2 = 0.025 from Z-table:

  * **Z = ±1.96**
* So, if computed Z lies **outside \[-1.96, 1.96]**, we **reject H₀**

#### 4. **Perform Z-Test**:

We would now use the Z formula:

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

$$
Z = \frac{169.5 - 168}{3.9 / \sqrt{36}} = \frac{1.5}{0.65} ≈ 2.31
$$

#### 5. **Decision**:

* Computed Z = **2.31**
* Since **2.31 > 1.96**, it lies **outside** the 95% confidence interval
* ✅ **Conclusion**: **Reject the Null Hypothesis** → There is enough evidence to suggest the mean is different from 168 cm.

---

### 🧠 **Key Concepts** Covered:

* Hypothesis structure (H₀ vs H₁)
* Conditions for using Z-Test
* Confidence Interval and Significance Level (α)
* Two-tailed testing
* Z-score calculation and interpretation


````python
import math
from scipy.stats import norm

# Given values
population_mean = 168
population_std = 3.9
sample_size = 36
sample_mean = 169.5
confidence_level = 0.95

# Step 1: Calculate standard error
standard_error = population_std / math.sqrt(sample_size)

# Step 2: Calculate Z-score
z_score = (sample_mean - population_mean) / standard_error

# Step 3: Determine critical Z-value for 95% confidence (two-tailed)
alpha = 1 - confidence_level
z_critical = norm.ppf(1 - alpha / 2)  # two-tailed

# Step 4: Decision
reject_null = abs(z_score) > z_critical

# Print results
print(f"Z-score: {z_score:.2f}")
print(f"Z-critical (two-tailed): ±{z_critical:.2f}")
print("Result:", "Reject null hypothesis" if reject_null else "Fail to reject null hypothesis")
````

````python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Define the mean and standard deviation
mu = 168  # population mean
sigma = 3.9  # population standard deviation
n = 36  # sample size
sample_mean = 169.5

# Standard error
se = sigma / np.sqrt(n)

# Compute Z score
z = (sample_mean - mu) / se

# Create x values for the plot
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, se)

# Critical z-values for 95% CI (two-tailed)
z_critical = 1.96
x_crit_left = mu - z_critical * se
x_crit_right = mu + z_critical * se

# Plot the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Normal Distribution", color="blue")

# Fill critical regions
plt.fill_between(x, y, where=(x <= x_crit_left), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= x_crit_right), color='red', alpha=0.5, label='Rejection Region (Right)')

# Plot sample mean as a vertical line
plt.axvline(sample_mean, color='green', linestyle='--', linewidth=2, label=f'Sample Mean = {sample_mean} cm\nZ = {z:.2f}')

# Plot critical boundaries
plt.axvline(x_crit_left, color='black', linestyle=':', label=f'Critical Value Left = {x_crit_left:.2f}')
plt.axvline(x_crit_right, color='black', linestyle=':', label=f'Critical Value Right = {x_crit_right:.2f}')

# Titles and labels
plt.title("Z-Test Visualization with Rejection Regions")
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.legend(loc="upper left")
plt.grid(True)

plt.tight_layout()
plt.show()
````

## 📊 **Student’s t-Distribution**

In this session, we discussed the **Student's t-distribution** and **when to use it** compared to the **z-distribution**.

---

### ✅ **When to Use z-distribution**

* Used when the **population standard deviation (σ)** is **known**
* Formula:

  $$
  Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
  $$

---

### 🚫 **The Problem with z-distribution**

* In many real-world scenarios, we **do not know the population standard deviation**
* It’s often **impossible or impractical** to calculate since we don’t have data for the entire population

---

### ✅ **When to Use Student's t-distribution**

* Use **t-distribution** when **σ is unknown**
* Replace σ with the **sample standard deviation (s)**
* New formula:

  $$
  t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
  $$

---

### 🧠 **Degree of Freedom (df)**

* Important parameter for the t-distribution
* Calculated as:

  $$
  \text{df} = n - 1
  $$
* Helps identify the correct row in the **t-table**

**Analogy:** If 3 people must choose 3 unique chairs, only the first 2 really have choices → df = 3 – 1 = 2

---

### 📈 **t-Table Overview**

* The t-table provides critical values based on:

  * **Degrees of freedom**
  * **Confidence level**
  * Whether it's a **one-tailed or two-tailed** test

## 📊 **Summary: One-Sample t-Test Explanation and Example**

### 🔹 **When to Use t-Test vs. z-Test**

* **Use z-test**: When **population standard deviation (σ)** is known and **n ≥ 30**
* **Use t-test**: When **σ is unknown** (we use **sample standard deviation** instead)

---

## 🧪 **Problem Statement**

> A new medication is tested to see if it affects intelligence.
> The **population IQ mean** is 100.
> A **sample of 30 people** who took the medication has a **mean IQ = 140** and **sample std dev = 20**.

---

## 📌 **Step-by-Step Approach**

### 1. **Known Values**

* Population mean (μ) = 100
* Sample mean ($\bar{x}$) = 140
* Sample std dev (s) = 20
* Sample size (n) = 30
* Confidence level = 95% → α = 0.05
* Degrees of freedom (df) = $n - 1 = 29$

---

### 2. **Hypotheses**

* **Null hypothesis (H₀):** μ = 100
* **Alternative hypothesis (H₁):** μ ≠ 100 (⇒ **two-tailed test**)

---

### 3. **Critical t-value from t-table**

* For **df = 29** and **α = 0.05 (two-tailed)**
* **Critical value (±t₀.₀₂₅)** ≈ **±2.045**

---

### 4. **Compute the Test Statistic**

$$
t = \frac{\bar{x} - \mu}{s / \sqrt{n}} = \frac{140 - 100}{20 / \sqrt{30}} ≈ 10.96
$$

---

### 5. **Decision Rule**

* If **t > 2.045** or **t < -2.045**, **reject H₀**

Since **t = 10.96 > 2.045**, it falls in the **rejection region**

---

## ✅ **Conclusion**

* The **null hypothesis is rejected**
* The medication **has a statistically significant effect** on IQ
* Since t is **positive and far right**, the medication **increased intelligence**

---

## 🏷️ Test Type:

**One-sample, two-tailed t-test**



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Given values
population_mean = 100
sample_mean = 140
sample_std = 20
sample_size = 30
confidence_level = 0.95
alpha = 1 - confidence_level
df = sample_size - 1

# Calculate t-statistic
standard_error = sample_std / np.sqrt(sample_size)
t_statistic = (sample_mean - population_mean) / standard_error

# Critical t value for two-tailed test
t_critical = t.ppf(1 - alpha/2, df)

# Create x range for t-distribution plot
x = np.linspace(-5, 5, 1000)
y = t.pdf(x, df)

# Plot the t-distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f't-distribution (df={df})')
plt.fill_between(x, y, where=(x <= -t_critical), color='red', alpha=0.3, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x >= t_critical), color='red', alpha=0.3, label='Rejection Region (right)')
plt.axvline(t_statistic, color='blue', linestyle='--', label=f't-statistic = {t_statistic:.2f}')
plt.axvline(-t_critical, color='black', linestyle=':', label=f'Critical value = ±{t_critical:.2f}')
plt.axvline(t_critical, color='black', linestyle=':')

plt.title('One-Sample Two-Tailed t-Test')
plt.xlabel('t-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Return t-statistic and decision
t_statistic, t_critical, abs(t_statistic) > t_critical

````

## 📊 **Summary: Type I and Type II Errors**

In this video, you explain the two types of errors that can occur during **hypothesis testing** — particularly when making decisions based on **null hypothesis (H₀)**.

---

## 🧠 **Two Key Aspects**

* **Reality**: Whether the null hypothesis is **actually true or false**
* **Decision**: What your **hypothesis test concludes** (accept or reject H₀)

---

## 📋 **All Possible Outcomes**

| **Reality**     | **Decision** (Test Result) | **Outcome**        | **Explanation**                    |
| --------------- | -------------------------- | ------------------ | ---------------------------------- |
| H₀ is **true**  | **Reject H₀**              | ❌ Type I Error     | You incorrectly rejected a true H₀ |
| H₀ is **false** | **Accept/Retain H₀**       | ❌ Type II Error    | You failed to reject a false H₀    |
| H₀ is **true**  | **Accept/Retain H₀**       | ✅ Correct Decision | You correctly accepted a true H₀   |
| H₀ is **false** | **Reject H₀**              | ✅ Correct Decision | You correctly rejected a false H₀  |

---

## 🔍 **Definitions**

* **Type I Error (False Positive)**

  * Rejecting H₀ when it is actually true
  * Controlled by **α (significance level)**

* **Type II Error (False Negative)**

  * Failing to reject H₀ when it is actually false
  * Controlled by **β (1 - Power)**

---

## 🧠 Real-World Analogy

| Example: Medical Test                                                    |
| ------------------------------------------------------------------------ |
| Type I Error: Diagnosing someone as sick when they are actually healthy. |
| Type II Error: Missing a disease in someone who is actually sick.        |

---


---

## 🔍 What is **β (beta)** in Hypothesis Testing?

* **β (beta)** is the **probability of making a Type II Error**.
* That means:
  👉 You **fail to reject the null hypothesis (H₀)** even though **H₀ is actually false**.

---

## 💡 What is **Power of a Test**?

* The **power** of a statistical test is defined as:

$$
\text{Power} = 1 - \beta
$$

* So, **Power** = Probability that you **correctly reject a false null hypothesis**.

---

## 📊 Quick Summary Table

| Concept               | Meaning                                                             |
| --------------------- | ------------------------------------------------------------------- |
| **Type II Error (β)** | Accepting H₀ when it’s actually false                               |
| **Power (1 - β)**     | Correctly rejecting H₀ when it’s actually false (desirable outcome) |

---

## ✅ Why is Power Important?

A **higher power** means:

* You’re **more likely to detect a true effect** when it exists.
* In research, we aim for **power ≥ 80%** (i.e., β ≤ 0.2)

---

## 🎯 How to Increase Power?

1. **Increase sample size**
2. Use a **higher significance level (α)** (e.g., from 0.01 to 0.05)
3. Reduce data variability
4. Use better measurement techniques

---

## 🧠 Real-Life Example

Suppose a new drug actually works:

* If your test fails to detect it (you conclude it doesn’t work), that’s a **Type II error (β)**
* But if your test **successfully detects** that the drug works, that’s **Power = 1 - β**

## 🎓 **Topic**: Bayes Statistics & Bayes’ Theorem

### 🔹 **1. What is Bayesian Statistics?**

* An approach to data analysis and parameter estimation using **Bayes' Theorem**.
* Helps in **updating probabilities** based on new evidence.

---

### 🔹 **2. Independent vs Dependent Events**

* **Independent Events**: One event does **not affect** another (e.g., rolling a die, tossing a coin).
* **Dependent Events**: One event **affects** the outcome of another (e.g., drawing marbles without replacement).

---

### 🔹 **3. Conditional Probability**

* Formula:

  $$
  P(A \cap B) = P(A) \cdot P(B|A)
  $$
* This is used when events are dependent.

---

### 🔹 **4. Bayes’ Theorem**

Derived from conditional probability:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Also written as:

$$
P(B|A) = \frac{P(A|B) \cdot P(B)}{P(A)}
$$

* **A, B**: Events
* **P(A|B)**: Probability of A **given** B is true
* **P(B|A)**: Probability of B **given** A is true
* **P(A), P(B)**: Individual probabilities (priors)

---

### 🔹 **5. Application in Machine Learning (Naive Bayes Classifier)**

Given features:

* `x₁ = size of house`, `x₂ = number of rooms`, `x₃ = location`
* `y = price (output)`

We compute:

$$
P(y | x₁, x₂, x₃) = \frac{P(y) \cdot P(x₁, x₂, x₃ | y)}{P(x₁, x₂, x₃)}
$$

Used to **predict `y`** given known features.

---

### ✅ **Key Takeaways**

* Bayes’ Theorem updates probability with **new evidence**.

## 📌 **Topic: Confidence Interval & Margin of Error**

### ✅ **What is a Confidence Interval (CI)?**

* A **confidence interval** gives a **range of values** likely to contain the **true population mean**.
* It's calculated using the formula:

  $$
  \text{CI} = \text{Point Estimate} \pm \text{Margin of Error}
  $$

---

### ✅ **When to Use Z-Test vs T-Test for CI**

* Use **Z-Test** when:

  * Population standard deviation (σ) is known.
* Use **T-Test** when:

  * Population standard deviation is unknown.
  * Sample size is small (n < 30).

---

### ✅ **Formula (Z-distribution)**

$$
\text{CI} = \bar{x} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$

Where:

* $\bar{x}$ = sample mean
* $\sigma$ = population standard deviation
* $n$ = sample size
* $Z_{\alpha/2}$ = z-score for desired confidence level (e.g., 1.96 for 95%)

---

### 🧪 **Example: CAT Verbal Section Scores**

* Population standard deviation (σ): **100**
* Sample mean ($\bar{x}$): **520**
* Sample size (n): **30**
* Confidence Level: **95%** (α = 0.05 → $Z_{\alpha/2} = 1.96$)

#### 📐 **Margin of Error:**

$$
\text{ME} = 1.96 \cdot \frac{100}{\sqrt{30}} ≈ 39.2
$$

#### 📍 **Confidence Interval:**

$$
\text{Lower Bound} = 520 - 39.2 = 480.8  
\text{Upper Bound} = 520 + 39.2 = 559.2
$$

---

### 📊 **Conclusion:**

> “We are 95% confident that the average CAT exam score lies between **480.8** and **559.2**.”

---

### 🔁 **Note:**

* If you use **T-distribution**, replace $Z_{\alpha/2}$ with $T_{\alpha/2, df}$, using degrees of freedom (n-1).
