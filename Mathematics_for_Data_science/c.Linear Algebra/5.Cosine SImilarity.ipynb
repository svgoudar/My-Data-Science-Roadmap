
* Dot product $= 5*0 + 0*2 = 0$
* **Perpendicular vectors**, hence **dot product zero**.

## 🚀 Application: Cosine similarity in recommendation systems

### 🎬 Example in data science

* Represent movies by feature vectors:

$$
\text{Avengers} = (1,2,0,3,1),\quad \text{Movie B} = (2,0,1,1,1)
$$

* Compute cosine similarity:

$$
\cos \theta = \frac{ \mathbf{a} \cdot \mathbf{b} }{||\mathbf{a}|| \times ||\mathbf{b}||}
$$

* Dot product:

$$
(1*2 + 2*0 + 0*1 + 3*1 + 1*1) = 2 + 0 + 0 + 3 + 1 = 6
$$

* Magnitudes:

$$
||a|| = \sqrt{1^2 + 2^2 + 0^2 + 3^2 + 1^2} = \sqrt{1+4+0+9+1}= \sqrt{15}=3.872
$$

$$
||b|| = \sqrt{2^2 + 0^2 + 1^2 + 1^2 + 1^2} = \sqrt{2^2+0+1+1+1} = \sqrt{7}=2.646
$$

* Cosine similarity:

$$
\cos \theta = \frac{6}{(3.872)(2.646)} = 0.586
$$

→ Movies are **58.6% similar**, can recommend **Movie B** to a user who liked **Avengers**.



```python
import numpy as np

def cosine_similarity(a, b):
    a = np.array(a)
    b = np.array(b)
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    cos_sim = dot_product / (norm_a * norm_b)
    
    print(f"Vector A: {a}")
    print(f"Vector B: {b}")
    print(f"Dot product: {dot_product}")
    print(f"||A|| (magnitude): {norm_a}")
    print(f"||B|| (magnitude): {norm_b}")
    print(f"Cosine similarity: {cos_sim:.4f}")
    
    return cos_sim

# Try different examples

print("\n--- Example 1: Similar direction ---")
cosine_similarity([5,0], [2,2])

print("\n--- Example 2: Opposite direction ---")
cosine_similarity([-5,0], [2,2])

print("\n--- Example 3: Perpendicular ---")
cosine_similarity([5,0], [0,2])

print("\n--- Example 4: Movie example ---")
cosine_similarity([1,2,0,3,1], [2,0,1,1,1])
```

```python
import matplotlib.pyplot as plt

def plot_vectors_with_cos_sim():
    origin = np.array([0,0])
    
    # Example vectors
    a1 = np.array([5,0])
    b1 = np.array([2,2])
    
    a2 = np.array([-5,0])
    b2 = np.array([2,2])
    
    a3 = np.array([5,0])
    b3 = np.array([0,2])
    
    plt.figure(figsize=(8,8))
    
    # Plot 1 - same direction
    plt.quiver(*origin, *a1, color='r', angles='xy', scale_units='xy', scale=1, label='a1 = (5,0)')
    plt.quiver(*origin, *b1, color='g', angles='xy', scale_units='xy', scale=1, label='b1 = (2,2)')
    cos_sim1 = cosine_similarity(a1, b1)
    plt.text(3,1, f"cos = {cos_sim1:.2f}", fontsize=12, color='purple')
    
    # Plot 2 - opposite
    plt.quiver(*origin, *a2, color='c', angles='xy', scale_units='xy', scale=1, label='a2 = (-5,0)')
    plt.quiver(*origin, *b2, color='m', angles='xy', scale_units='xy', scale=1, label='b2 = (2,2)')
    cos_sim2 = cosine_similarity(a2, b2)
    plt.text(-3,1, f"cos = {cos_sim2:.2f}", fontsize=12, color='purple')
    
    # Plot 3 - perpendicular
    plt.quiver(*origin, *a3, color='orange', angles='xy', scale_units='xy', scale=1, label='a3 = (5,0)')
    plt.quiver(*origin, *b3, color='blue', angles='xy', scale_units='xy', scale=1, label='b3 = (0,2)')
    cos_sim3 = cosine_similarity(a3, b3)
    plt.text(3,2, f"cos = {cos_sim3:.2f}", fontsize=12, color='purple')
    
    plt.xlim(-6,6)
    plt.ylim(-1,6)
    plt.grid(True)
    plt.legend()
    plt.title("Cosine Similarity: Same, Opposite, Perpendicular Vectors")
    plt.show()

plot_vectors_with_cos_sim()
```

### 🚀 What is RAG?

* **RAG** stands for **Retrieval Augmented Generation**.
* It combines **retrieval from a knowledge base** (like a database of documents or books) with **generative models** (like LLMs) to produce richer, more accurate responses.

---

### ⚙️ Example use case

* Suppose you have a **book of 500 pages**.
* You want to build a **chatbot** so that when a user asks:

  > "What is physics?"
  > the chatbot retrieves and responds with the answer **from that book**, without the user reading it.

---

### 🧠 How does this work?

1️⃣ **Convert book content to vectors**

* All sentences, paragraphs or chunks from the book are converted to **vectors (embeddings)**.
* Methods include:

  * Bag of Words (BoW)
  * TF-IDF
  * Word2Vec
  * Modern embeddings (like OpenAI / HuggingFace sentence transformers)

2️⃣ **Store these vectors in a vector database**

* Similar to MySQL/SQL Server for structured data.
* Examples: Pinecone, Chroma, Milvus, FAISS.

3️⃣ **When a user asks a question**

* The question is converted into a **query vector** using the **same embedding method**.

4️⃣ **Cosine similarity search**

* The system performs a **similarity search** in the vector DB.
* It finds the vector(s) from the book with **highest cosine similarity** to the query vector.

5️⃣ **Retrieve & respond**

* The most similar text chunks are returned.
* They are converted back to text and displayed to the user.

---

### 🎯 Why is this powerful?

* Enables answering questions **without reading the entire book**.
* Powers modern AI chatbots that can answer from:

  * PDFs
  * Videos (transcripts)
  * Websites
  * Internal company docs.

---

### 🔑 Key takeaway

> Even advanced GenAI systems like RAG-powered chatbots **ultimately rely on fundamental vector operations**, especially **cosine similarity**, to find and return relevant content.


```python
import numpy as np
import faiss

# -------------------------
# Example data (simulated embeddings)
# -------------------------
# Let's say these are sentence embeddings (5 documents, each 4-dimensional embedding)
documents = np.array([
    [0.1, 0.3, 0.5, 0.7],
    [0.2, 0.2, 0.2, 0.2],
    [0.9, 0.1, 0.3, 0.4],
    [0.0, 0.1, 0.2, 0.9],
    [0.3, 0.8, 0.1, 0.2],
]).astype('float32')

# -------------------------
# Normalize for cosine similarity
# -------------------------
# FAISS uses L2 distance. To do cosine similarity, normalize vectors.
faiss.normalize_L2(documents)

# -------------------------
# Build FAISS index
# -------------------------
dimension = documents.shape[1]
index = faiss.IndexFlatIP(dimension)  # inner product is equivalent to cosine sim on normalized
index.add(documents)

# -------------------------
# Query vector
# -------------------------
query = np.array([[0.1, 0.2, 0.3, 0.9]], dtype='float32')
faiss.normalize_L2(query)

# -------------------------
# Search: get top 3 most similar vectors
# -------------------------
k = 3
distances, indices = index.search(query, k)

# -------------------------
# Show results
# -------------------------
print("Query:", query)
for i, (idx, score) in enumerate(zip(indices[0], distances[0])):
    print(f"Rank {i+1}: Document #{idx} with cosine similarity {score:.4f}")
```

```python
import numpy as np
import faiss
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# -----------------------
# Example embeddings
# -----------------------
documents = np.array([
    [0.1, 0.3, 0.5, 0.7],
    [0.2, 0.2, 0.2, 0.2],
    [0.9, 0.1, 0.3, 0.4],
    [0.0, 0.1, 0.2, 0.9],
    [0.3, 0.8, 0.1, 0.2],
]).astype('float32')

query = np.array([[0.1, 0.2, 0.3, 0.9]], dtype='float32')

# -----------------------
# Normalize for cosine sim
# -----------------------
faiss.normalize_L2(documents)
faiss.normalize_L2(query)

# -----------------------
# Build index & search
# -----------------------
dimension = documents.shape[1]
index = faiss.IndexFlatIP(dimension)
index.add(documents)

k = 3
distances, indices = index.search(query, k)

# -----------------------
# Reduce to 2D for plot
# -----------------------
pca = PCA(n_components=2)
all_data = np.vstack([documents, query])
data_2d = pca.fit_transform(all_data)
docs_2d = data_2d[:-1]
query_2d = data_2d[-1]

# -----------------------
# Plot
# -----------------------
plt.figure(figsize=(8,8))
plt.scatter(docs_2d[:,0], docs_2d[:,1], c='blue', label='Document vectors')
plt.scatter(query_2d[0], query_2d[1], c='red', marker='X', s=100, label='Query vector')

# Highlight top-k
for rank, idx in enumerate(indices[0]):
    plt.scatter(docs_2d[idx,0], docs_2d[idx,1], edgecolors='gold', facecolors='none', s=300, linewidths=2)
    plt.text(docs_2d[idx,0]+0.01, docs_2d[idx,1]+0.01, f'Rank {rank+1}', fontsize=10, color='darkorange')

plt.title("FAISS Cosine Similarity Search (2D PCA projection)")
plt.legend()
plt.grid(True)
plt.show()
```
