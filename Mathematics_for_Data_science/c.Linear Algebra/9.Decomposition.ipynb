# üöÄ What is **matrix decomposition**?

Matrix decomposition (also called **matrix factorization**) means **breaking down a matrix into a product of simpler, special matrices**.

It‚Äôs like saying:

$$
\text{Matrix} = \text{(transform 1)} \times \text{(transform 2)} \times \text{...}
$$

This is **hugely useful** because:

* simpler matrices are easier to analyze, compute inverses, solve systems, or understand geometry.
* many machine learning and data algorithms (like PCA, recommender systems, or least squares) rely on decomposing matrices.

---

# üîç Examples of matrix decompositions

| Decomposition | Factorization form     | Works for                   | Use cases                                  |
| ------------- | ---------------------- | --------------------------- | ------------------------------------------ |
| **LU**        | $A = L U$              | square matrices             | solving $Ax = b$, inverses                 |
| **QR**        | $A = Q R$              | any $m \times n$            | least squares, orthonormal bases           |
| **SVD**       | $A = U S V^T$          | any $m \times n$            | PCA, dimensionality reduction, compression |
| **Cholesky**  | $A = L L^T$            | symmetric positive definite | fast solves, covariance sampling           |
| **Eigen**     | $A = V \Lambda V^{-1}$ | diagonalizable matrices     | spectral analysis, dynamical systems       |

---

# üöÄ Intuition

It‚Äôs like taking a complex transformation (matrix multiplication), and rewriting it as:

* a rotation or reflection (**orthogonal matrix**),
* a scaling (**diagonal matrix**),
* another rotation.

That‚Äôs what SVD does:

$$
A = U S V^T
$$

where $U$ and $V$ are rotations, and $S$ is scaling.

---

# üéØ Why is it important in data science & ML?

‚úÖ Because many problems become **trivial** once decomposed:

* Solve $Ax = b$ efficiently (via LU / QR instead of costly inverses).
* Find **principal components** (SVD is the engine behind PCA).
* Regularize or compress models (low-rank approximations via SVD).
* Sample from multivariate distributions (via Cholesky).

# üöÄ 1Ô∏è‚É£ QR decomposition

### üîç What is QR?

It decomposes a matrix $A$ into:

$$
A = Q R
$$

where:

* **$Q$** is an **orthogonal (or unitary)** matrix (meaning $Q^T Q = I$).
* **$R$** is an **upper triangular** matrix.

### üìå Intuition

It‚Äôs like breaking down the matrix into a rotation/reflection part (**Q**) and a stretch/shear part (**R**).

### ‚öôÔ∏è Use cases

* Solving **least squares problems** (linear regression normal equations).
* Finding **orthonormal bases** (for PCA / Gram-Schmidt).
* More numerically stable than directly computing $(A^T A)^{-1} A^T$.

### üêç Example

```python
Q, R = np.linalg.qr(A)
```

`Q` is orthogonal, `R` is upper triangular.

---

# üöÄ 2Ô∏è‚É£ Singular Value Decomposition (SVD, U, S, Vt)

### üîç What is SVD?

Decomposes $A$ into three matrices:

$$
A = U S V^T
$$

where

* **$U$** contains the **left singular vectors** (orthogonal).
* **$S$** is a diagonal matrix with **singular values** (stretching factors).
* **$V^T$** (or $V^H$) has the **right singular vectors**.

### üìå Intuition

It‚Äôs like rotating to align with the principal axes, scaling (via singular values), then rotating again.

### ‚öôÔ∏è Use cases

* Dimensionality reduction (PCA is based on SVD).
* Solving **ill-conditioned systems**.
* Low-rank approximations (compress images/data).

### üêç Example

```python
U, S, Vt = np.linalg.svd(A)
```

* `U` gives directions in original space.
* `S` gives strength along those directions.
* `Vt` gives directions in input space.

---

# üöÄ 3Ô∏è‚É£ Cholesky decomposition

### üîç What is Cholesky?

For **symmetric positive-definite matrices** $A$,

$$
A = L L^T
$$

where $L$ is a **lower triangular matrix**.

### üìå Intuition

Like a **square-root of a matrix**, easier to work with than full inverse.

### ‚öôÔ∏è Use cases

* Faster than general `np.linalg.inv` for special matrices.
* Used in sampling from multivariate normals (covariance matrix factorization).
* Solving linear systems: solve $L y = b$ then $L^T x = y$.

### üêç Example

```python
L = np.linalg.cholesky(A)
```

(works only if $A$ is symmetric positive definite).

| **SVD**       | $A = U S V^T$ | PCA, compression, stable inversion             |
| **Cholesky**  | $A = L L^T$   | Fast solving of systems, multivariate sampling |

````python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# 1. Create toy data
# -------------------------
np.random.seed(0)
X = np.random.multivariate_normal([0, 0], [[3, 2.5], [2.5, 2]], 200)

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], alpha=0.5)
plt.axis('equal')
plt.title("Original data")
plt.show()

# -------------------------
# 2. Center data
# -------------------------
X_mean = np.mean(X, axis=0)
X_centered = X - X_mean

# -------------------------
# 3. PCA via Covariance + Eigen decomposition
# -------------------------
cov_matrix = np.cov(X_centered.T)  # shape (2,2)
eigvals, eigvecs = np.linalg.eig(cov_matrix)

# Sort by eigenvalues
order = np.argsort(eigvals)[::-1]
eigvals = eigvals[order]
eigvecs = eigvecs[:, order]

# Plot principal components
plt.figure(figsize=(6,6))
plt.scatter(X_centered[:,0], X_centered[:,1], alpha=0.3)
origin = np.mean(X_centered, axis=0)

for i in range(2):
    plt.quiver(*origin, eigvecs[0,i]*eigvals[i], eigvecs[1,i]*eigvals[i], 
               angles='xy', scale_units='xy', scale=1, color='r')
plt.axis('equal')
plt.title("PCA via Eigen decomposition")
plt.show()

# -------------------------
# 4. PCA via SVD
# -------------------------
U, S, Vt = np.linalg.svd(X_centered)
# Vt[0] is the first principal component

plt.figure(figsize=(6,6))
plt.scatter(X_centered[:,0], X_centered[:,1], alpha=0.3)
plt.quiver(*origin, Vt[0,0]*S[0], Vt[0,1]*S[0], 
           angles='xy', scale_units='xy', scale=1, color='g')
plt.quiver(*origin, Vt[1,0]*S[1], Vt[1,1]*S[1], 
           angles='xy', scale_units='xy', scale=1, color='b')
plt.axis('equal')
plt.title("PCA via SVD")
plt.show()

# -------------------------
# 5. Dimensionality reduction: project onto first PC
# -------------------------
X_pca_eigen = X_centered @ eigvecs[:, 0]
X_pca_svd = U[:,0] * S[0]

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(X_pca_eigen, np.zeros_like(X_pca_eigen), alpha=0.5)
plt.title("Projected data (Eigen PCA)")

plt.subplot(1,2,2)
plt.scatter(X_pca_svd, np.zeros_like(X_pca_svd), alpha=0.5)
plt.title("Projected data (SVD PCA)")
plt.show()
````

# üßÆ 1. Eigen decomposition (Eigenvalues & Eigenvectors)

---

### üí° Intuition

* Eigen decomposition breaks down a **square matrix $A$** into its **directions of stretching** (eigenvectors) and **how much it stretches them** (eigenvalues).

* For a square matrix $A$:

$$
A v = \lambda v
$$

where

* $v$ = eigenvector (direction)
* $\lambda$ = eigenvalue (scale factor)

This means multiplying $A$ by $v$ simply scales it by $\lambda$.

---

### üìê Matrix form

If $A$ is **diagonalizable**, then:

$$
A = V \Lambda V^{-1}
$$

where:

* $V$ = matrix of eigenvectors (columns are eigenvectors)
* $\Lambda$ = diagonal matrix with eigenvalues on the diagonal.

---

### üîç NumPy example

```python
import numpy as np

A = np.array([[4, -2],
              [1,  1]])

# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Reconstruct A
Lambda = np.diag(eigenvalues)
V = eigenvectors
V_inv = np.linalg.inv(V)
A_reconstructed = V @ Lambda @ V_inv
print("Reconstructed A:\n", A_reconstructed)
```

# üèóÔ∏è 2. QR decomposition

---

### üí° Intuition

* QR breaks a matrix into:

$$
A = Q R
$$

where

* $Q$ = orthogonal (or unitary) matrix (columns are **orthonormal vectors**),
* $R$ = upper triangular matrix.

This is very useful for solving linear systems and for finding eigenvalues (via QR algorithm).

Think of $Q$ as rotating the space, and $R$ as stretching it along new axes.

---

### üìê Matrix form

For any **square or tall matrix $A$** (shape $m \times n$, $m \ge n$):

$$
A = Q R
$$

where:

* $Q$ is $m \times n$, orthonormal columns,
* $R$ is $n \times n$, upper triangular.

---

### üîç NumPy example

```python
import numpy as np

A = np.array([[12, -51, 4],
              [6, 167, -68],
              [-4, 24, -41]])

Q, R = np.linalg.qr(A)

print("Q:\n", Q)
print("R:\n", R)

# Reconstruct A
print("Q @ R:\n", Q @ R)
```

# üî¨ 3. U decomposition (actually SVD)

---

### üí° Intuition

This is usually called **SVD (Singular Value Decomposition)**, where

$$
A = U \Sigma V^T
$$

* $U$: left singular vectors (orthogonal)
* $\Sigma$: diagonal matrix of singular values (like eigenvalues for general rectangular matrices)
* $V$: right singular vectors (orthogonal)

SVD generalizes eigen-decomposition to **rectangular matrices**.

* Great for compression (PCA), noise reduction.

---

### üìê Matrix form

For any matrix $A$ (shape $m \times n$):

$$
A = U \Sigma V^T
$$

* $U$ ($m \times m$): orthogonal
* $\Sigma$ ($m \times n$): diagonal (only diagonal elements nonzero)
* $V$ ($n \times n$): orthogonal

---

### üîç NumPy example


````python
import numpy as np

A = np.array([[3, 1, 1],
              [-1, 3, 1]])

U, S, VT = np.linalg.svd(A)

print("U:\n", U)
print("Singular values S:\n", S)
print("V^T:\n", VT)

# Reconstruct A
Sigma = np.zeros((A.shape[0], A.shape[1]))
print(Sigma)
for i in range(len(S)):
    Sigma[i, i] = S[i]
A_reconstructed = U @ Sigma @ VT
print("Reconstructed A:\n", A_reconstructed)
````
