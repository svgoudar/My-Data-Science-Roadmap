## üìå **2. What is a function (in linear algebra context)?**

* **Definition:**
  A function is a mathematical relationship that uniquely **maps elements from one set (domain)** to **another set (codomain)**.

* **Notation:**

  $$
  f : X \rightarrow Y
  $$

  Meaning: function $f$ maps set $X$ (domain) to set $Y$ (codomain).

* If $x$ is an element in $X$, then $f(x)$ gives the corresponding element in $Y$.

---

## üìù **3. Simple example with real numbers**

* Given:

  $$
  f(x) = 2x + 3
  $$
* Maps a real number $x$ to another real number $y$.
* E.g.,

  * If $x = 2$,

  $$
  f(2) = 2 \times 2 + 3 = 7
  $$
* This is a mapping from $x \in \mathbb{R}$ to $y \in \mathbb{R}$.

## üîÑ **4. Example with vectors (mapping dimensions)**

* Suppose we have a 3D vector:

$$
\mathbf{v} = (x, y, z) \in \mathbb{R}^3
$$

* Define a function:

$$
f(\mathbf{v}) = (x + y, 6z)
$$

* This **maps the 3D vector to a 2D vector**, i.e.:

$$
f : \mathbb{R}^3 \rightarrow \mathbb{R}^2
$$

* This is an example of **reducing dimensionality through a function** (transformation).

### üöÄ What is a linear transformation?

* A **linear transformation** is a function between two vector spaces that preserves:

  1. **Vector addition:**

     $$
     T(u + v) = T(u) + T(v)
     $$
  2. **Scalar multiplication:**

     $$
     T(c \cdot u) = c \cdot T(u)
     $$
* It maps vectors from one space (domain) to another (codomain) while maintaining these properties.

---

### üìù Formal conditions

* If $T: V \to W$ is a linear transformation, then for all vectors $u, v \in V$ and scalars $c$,

  * **Additivity:**

    $$
    T(u + v) = T(u) + T(v)
    $$
  * **Homogeneity:**

    $$
    T(c \cdot u) = c \cdot T(u)
    $$

---

### ‚úÖ Example proving a transformation is linear

#### Reflection across the **y-axis**

* Defined by:

  $$
  T(x, y) = (-x, y)
  $$
* Expressed using a matrix:

  $$
  A = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}
  $$
* Then:

  $$
  T(\mathbf{x}) = A \cdot \mathbf{x}
  $$

#### üîç Checking the properties

1. **Additivity:**
   Verified that:

   $$
   T(u + v) = T(u) + T(v)
   $$

   by direct matrix multiplication and addition.
2. **Homogeneity:**
   Verified that:

   $$
   T(c \cdot u) = c \cdot T(u)
   $$

   by multiplying first by scalar then applying transformation vs applying transformation first then multiplying.

‚úÖ Conclusion: **Reflection is a linear transformation.**

---

### ‚ùå Example of a transformation that is *not* linear

#### Transformation with a fixed vector added

* Defined by:

  $$
  T(x) = x + b
  $$

  where $b = (1,1)$ is a fixed vector.

#### üîç Checking the properties

1. **Additivity:**

   * Calculated $T(u+v) \neq T(u) + T(v)$
2. **Homogeneity:**

   * Calculated $T(cu) \neq cT(u)$

‚ùå Conclusion: **This is *not* a linear transformation** because it fails both additivity and homogeneity.

### üîç Key points explained

#### üöÄ What is the magnitude (vector length)?

* It is the **distance from the origin** to the point represented by the vector.
* For a vector $\mathbf{x} = (x_1, x_2, ..., x_n)$, the magnitude is:

$$
\| \mathbf{x} \| = \sqrt{x_1^2 + x_2^2 + ... + x_n^2}
$$

---

#### ‚úç Example in 2D

* Given vector $\mathbf{a} = (2, 3)$:

  * By Pythagoras theorem:

$$
\| \mathbf{a} \| = \sqrt{2^2 + 3^2} = \sqrt{4 + 9} = \sqrt{13}
$$

---

#### üõ† What is a unit vector?

* A **unit vector** is a vector of **length 1** pointing in the same direction as the original vector.

* Denoted by a **hat**, e.g. $\hat{u}$ (read as "u cap").

---

#### ‚öô How to calculate a unit vector?

* Given any vector $\mathbf{v}$, the unit vector $\hat{u}$ in the same direction is:

$$
\hat{u} = \frac{\mathbf{v}}{\|\mathbf{v}\|}
$$

* That means you **divide each component by the magnitude**, normalizing it to length 1.

---

#### ‚úç Example in 3D

* Given $\mathbf{v} = (1, 2, 0)$:

1. Compute magnitude:

$$
\| \mathbf{v} \| = \sqrt{1^2 + 2^2 + 0^2} = \sqrt{5}
$$

2. Compute unit vector:

$$
\hat{u} = \frac{1}{\sqrt{5}}(1,2,0) = \left(\frac{1}{\sqrt{5}}, \frac{2}{\sqrt{5}}, 0\right)
$$

* Verification:

$$
\| \hat{u} \| = \sqrt{\left(\frac{1}{\sqrt{5}}\right)^2 + \left(\frac{2}{\sqrt{5}}\right)^2 + 0^2} 
= \sqrt{\frac{1}{5} + \frac{4}{5}} 
= \sqrt{1} = 1
$$

---

### üß≠ Why is this important?

* **Normalization**:

  * Converts vectors to unit vectors, ensuring all have length 1.
  * Useful for machine learning algorithms, improves convergence and optimization.
* E.g. scaling data to range \[0,1] or \[-1,1] helps gradient descent optimize faster.

---

### üìå Connection to coordinate notation

* In 2D:

  * $i$ and $j$ (or $\hat{i}$ and $\hat{j}$) are **unit vectors along x and y**.
  * E.g. $(2,3) = 2\hat{i} + 3\hat{j}$.

---

‚úÖ **In short:**

> * The **magnitude** measures the length of the vector.
> * A **unit vector** has length 1 and points in the same direction.
> * Unit vectors are critical for normalization, helping machine learning models perform better.



```python
import numpy as np
import matplotlib.pyplot as plt

# Original vector
v = np.array([2, 3])

# Compute magnitude
magnitude_v = np.linalg.norm(v)

# Compute unit vector
u = v / magnitude_v

# Plot settings
fig, ax = plt.subplots(figsize=(6,6))
ax.set_xlim(-1, 3)
ax.set_ylim(-1, 4)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.5)

# Plot original vector
ax.arrow(0, 0, v[0], v[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue', label='Vector v')

# Plot unit vector (scaled for visibility)
ax.arrow(0, 0, u[0], u[1], head_width=0.1, head_length=0.1, fc='red', ec='red', label='Unit Vector u')

# Annotate
ax.text(v[0]/2, v[1]/2, r'$\mathbf{v}$', fontsize=14, color='blue')
ax.text(u[0]/2, u[1]/2, r'$\hat{\mathbf{u}}$', fontsize=14, color='red')

# Legend
ax.legend(['Original vector v', 'Unit vector u'])

plt.title('Vector and Unit Vector')
plt.show()
```

## Projection

* **Projection** means dropping (or "casting") a vector onto a line (or another vector direction), finding **how much of that vector lies along the line**.
* Mathematically, it‚Äôs like finding the **shadow of a vector** on a line.

---

### ‚úçÔ∏è The key formula

If you want to project vector **x** onto line **L**, which is defined by vector **v**,
the formula for the projection is:

$$
\text{proj}_{L}(x) = \left(\frac{x \cdot v}{v \cdot v}\right) v
$$

This gives a **new vector along line L**, telling how much of **x** lies in the direction of **v**.

---

### üîç Small numerical example

* Line defined by vector $v = (2, 1)$
* Vector to project: $x = (2, 3)$

Calculate:

$$
c = \frac{x \cdot v}{v \cdot v} = \frac{(2 \times 2 + 3 \times 1)}{(2 \times 2 + 1 \times 1)} = \frac{7}{5}
$$

So projection is:

$$
\text{proj}_L(x) = c v = \frac{7}{5} (2,1) = \left(2.8, 1.4\right)
$$

---
---

### ‚úÖ Important geometric intuition

* The **projection vector always lies on line L**.
* The **difference** between $x$ and its projection is **perpendicular** to $L$.
* If two vectors are **perpendicular**, their **dot product is zero**.

---

### üéØ Why is this useful?

* Projections help in breaking vectors into **components along directions**, which is critical in:

  * least squares fitting
  * decomposing forces in physics
  * graphics and computer vision
  * orthogonal projections in PCA


```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
x = np.array([2, 3])
v = np.array([2, 1])

# Compute projection
c = np.dot(x, v) / np.dot(v, v)
proj = c * v

# Plotting
plt.figure(figsize=(8, 8))
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)

# Original vector x
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'$\vec{x}$ (2,3)')

# Line defined by vector v (showing multiple to look like a line)
for t in np.linspace(-3, 3, 10):
    plt.plot([t * v[0]], [t * v[1]], 'go', alpha=0.5)

# Projection vector
plt.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'$\mathrm{proj}_L(\vec{x})$ (2.8,1.4)')

# Draw dashed line from tip of x to projection
plt.plot([x[0], proj[0]], [x[1], proj[1]], 'k--', linewidth=1)

plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.title("Projection of vector $\\vec{x}$ onto line $L$ defined by $\\vec{v}$")
plt.show()
```

## üöÄ Inverse and Identity Functions

### üîÑ What is an inverse function?

* Given a function **f** mapping elements from set **X** to set **Y** (i.e., f(x) = y),
* The **inverse function** (written as **f‚Åª¬π**) reverses this mapping, taking elements from **Y** back to **X**.
* That means:

  * **f: X ‚Üí Y**
  * **f‚Åª¬π: Y ‚Üí X**

### üìù Formal conditions

* For a function to have an inverse, it must satisfy:

  1. **f‚Åª¬π(f(x)) = x**, for all **x** in **X**.
  2. **f(f‚Åª¬π(y)) = y**, for all **y** in **Y**.

### üéØ Key requirement: Bijectiveness

* A function **f** has an inverse **iff** it is **bijective**, meaning:

  * **Injective (1-1):** Different elements in domain map to different elements in codomain.
  * **Surjective (onto):** Every element in codomain is mapped by some element from domain.

---

## ‚ö° Example of finding inverse

* Given **f(x) = 2x + 3**, find inverse:

  * Write **y = 2x + 3**, solve for **x**:

    $$
    x = \frac{y - 3}{2}
    $$
  * Thus, inverse function is:

    $$
    f^{-1}(y) = \frac{y - 3}{2}
    $$

---

## ‚úÖ Verification

* Check composition:

  * **f(f‚Åª¬π(y)) = 2((y-3)/2)+3 = y**
  * **f‚Åª¬π(f(x)) = (2x+3-3)/2 = x**

---

## ü™û What is an identity function?

* **Identity function I(x)** maps any element to itself:

  $$
  I(x) = x
  $$
* In matrix terms, **identity matrix** is a square matrix with 1s on diagonal and 0s elsewhere.

  * e.g., for 2x2:

    $$
    I = \begin{bmatrix}1 & 0\\0 & 1\end{bmatrix}
    $$
* Applying identity transformation returns the same vector.

---

## ‚ú® Properties of identity function

1. **Preservation:** Does not change any element.

   $$
   I(x) = x
   $$
2. **Linearity:**

   * **Additivity:**

     $$
     I(u+v) = I(u) + I(v)
     $$
   * **Homogeneity:**

     $$
     I(cu) = cI(u) = cu
     $$
3. **It‚Äôs its own inverse:**

   $$
   I(I(x)) = I(x) = x
   $$

---

## üöÄ Applications

* Inverse functions are crucial in data science & ML, e.g.:

  * **Scaling back normalized data.**
  * **Solving systems of equations.**
  * **Matrix inversions in algorithms like linear regression.**
* Identity functions help:

  * In linear algebra proofs.
  * As base cases in transformations.


```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its inverse
def f(x):
    return 2*x + 3

def f_inverse(y):
    return (y - 3) / 2

# Generate values
x = np.linspace(-10, 10, 400)
y = f(x)
y_vals = np.linspace(-20, 20, 400)
x_inv = f_inverse(y_vals)

# Plot
plt.figure(figsize=(10,6))
plt.plot(x, y, label=r"$f(x) = 2x+3$", color='blue')
plt.plot(x_inv, y_vals, label=r"$f^{-1}(y) = \frac{y-3}{2}$", color='green')
plt.plot(x, x, '--', color='gray', label='y = x (identity line)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Its Inverse')
plt.legend()
plt.grid(True)
plt.show()
```

### üü¢ **Standardization (Z-score normalization)**

* Formula:

  $$
  z = \frac{x - \mu}{\sigma}
  $$
* Transforms data to **mean = 0, std = 1** (standard normal distribution).
* Example: Number of rooms `[1,2,3,4]` becomes standardized to smaller centered values.

#### üîÑ Inverse transformation

* To get back to original scale:

$$
x = z \cdot \sigma + \mu
$$

* Used to **interpret model predictions** back in original units.

---

### üîµ **Normalization (Min-Max scaling)**

* Formula:

$$
z = \frac{x - \min(x)}{\max(x) - \min(x)}
$$

* Scales data to **\[0,1]** range.

#### üîÑ Inverse transformation

$$
x = z \cdot (\max(x) - \min(x)) + \min(x)
$$

* Common in **computer vision & image data.**

---

### üöÄ **Why inverse needed?**

* After ML predictions on scaled data, we need **inverse scaling** to convert back to original units for interpretation (like INR, sqft, etc).

1. **Determinant of a matrix**

   * A scalar value that tells:

     * If the matrix is invertible (non-zero determinant).
     * Geometric meaning: scales area (or volume in higher dims) under linear transformation.

2. **Inverse of a matrix**

   * For a 2x2 matrix

     $$
     A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}
     $$
   * The inverse is:

     $$
     A^{-1} = \frac{1}{\det(A)} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
     $$
   * Where

     $$
     \det(A) = ad - bc
     $$

3. **Verification using vector transformation**

   * Given $Y = A X$, we can recover $X$ using:

     $$
     X = A^{-1} Y
     $$
   * If calculation is correct, applying $A^{-1}$ should get back the original vector.

---

## üöÄ **Python code: Compute inverse, determinant & verify**


```python
import numpy as np

# Given matrix A
A = np.array([[4, 7],
              [2, 6]])

# Step 1: Compute determinant
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

# Check invertibility
if det_A != 0:
    print("Matrix A is invertible.")
else:
    raise ValueError("Matrix is not invertible.")

# Step 2: Compute inverse manually
A_inv_manual = (1/det_A) * np.array([[ A[1,1], -A[0,1]],
                                     [-A[1,0],  A[0,0]]])
print("Inverse of A (manual):\n", A_inv_manual)

# Or use numpy directly
A_inv_np = np.linalg.inv(A)
print("Inverse of A (numpy):\n", A_inv_np)

# Step 3: Verify with a vector transformation
x = np.array([[1],
              [1]])

# Apply transformation: y = A * x
y = A @ x
print("Transformed vector y:", y.flatten())

# Recover x: x = A_inv * y
x_recovered = A_inv_manual @ y
print("Recovered x:", x_recovered.flatten())

# Should be close to original x
assert np.allclose(x, x_recovered), "Recovery failed"
print("Recovery successful: x_recovered is close to original x.")
```
