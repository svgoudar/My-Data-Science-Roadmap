### üéØ **Vectors**

* **Definition (in physics):** A quantity having both **magnitude and direction**.
* **Definition (in data science):** An **ordered list of numbers** (multi-dimensional data), representing a point in space or record with multiple features.

    * E.g. student `[IQ, study_hours] = [90, 3]` is a 2D vector.
    * The weights of a person over time `[70, 72, 75, 73]` ‚Üí a 4D vector.

‚úÖ **Sample plot: vector addition in 2D**

Let‚Äôs create a plot for your example:

* $p_1 = (-4, 3)$
* $p_2 = (5, 3)$
* $p_1 + p_2 = (1, 6)$

````python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors
origin = np.array([0, 0])
p1 = np.array([4, 3])
p2 = np.array([5, 3])
p_sum = p1 + p2  # (9,6)

plt.figure(figsize=(8,8))

# Plot vector p1 from origin
plt.quiver(*origin, *p1, angles='xy', scale_units='xy', scale=1, color='r', label='p1 = (4,3)')

# Plot vector p2 starting from p1 (tip-to-tail)
plt.quiver(*p1, *p2, angles='xy', scale_units='xy', scale=1, color='g', label='p2 = (5,3) from p1')

# Plot resultant vector from origin to p_sum
plt.quiver(*origin, *p_sum, angles='xy', scale_units='xy', scale=1, color='b', label='p1 + p2 = (9,6)')

# Optional: draw dashed line to show the triangle
plt.plot([p1[0], p_sum[0]], [p1[1], p_sum[1]], 'b--')

# Set plot limits and grid
plt.xlim(-1, 10)
plt.ylim(-1, 8)
plt.grid(True)
plt.legend()
plt.title("Vector Addition in 2D")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
````

## üéØ Dot Product

### üîé Definition

The **dot product** of two vectors is the **sum of the products of their corresponding components**, resulting in a **scalar**.

$$
\mathbf{a} \cdot \mathbf{b} = a_1b_1 + a_2b_2 + ... + a_nb_n
$$

---

### üìä Example calculation

For vectors:

$$
\mathbf{a} = (2,3), \quad \mathbf{b} = (4,5)
$$

Dot product:

$$
(2 \times 4) + (3 \times 5) = 8 + 15 = 23
$$

---

## üìê Geometric interpretation

### üîπ Projection view

The dot product can be viewed as:

$$
\mathbf{a} \cdot \mathbf{b} = |\mathbf{a}| \times |\mathbf{b}| \times \cos(\theta)
$$

where $\theta$ is the angle between the vectors.

---

## üñº Plots to visualize dot product

### ‚û°Ô∏è 1. Same direction (positive dot product)


````python
import matplotlib.pyplot as plt
import numpy as np

origin = np.array([0,0])
a = np.array([5,0])
b = np.array([2,2])

plt.figure(figsize=(6,6))
plt.quiver(*origin, *a, color='r', scale=1, scale_units='xy', angles='xy', label='a = (5,0)')
plt.quiver(*origin, *b, color='g', scale=1, scale_units='xy', angles='xy', label='b = (2,2)')
plt.grid()
plt.axis('equal')
plt.legend()
plt.title("Dot product positive: vectors pointing roughly same way")
plt.show()

````


* Here dot product $= 5*2 + 0*2 = 10$
* Both vectors point **roughly in the same direction**, hence **positive dot product**.

### ‚û°Ô∏è 2. Opposite direction (negative dot product)



````python
import matplotlib.pyplot as plt
import numpy as np
a = np.array([-5,0])
b = np.array([2,2])

plt.figure(figsize=(6,6))
plt.quiver(*origin, *a, color='r', scale=1, scale_units='xy', angles='xy', label='a = (-5,0)')
plt.quiver(*origin, *b, color='g', scale=1, scale_units='xy', angles='xy', label='b = (2,2)')
plt.grid()
plt.axis('equal')
plt.legend()
plt.title("Dot product negative: vectors opposite directions")
plt.show()
````



* Dot product $= (-5)*2 + 0*2 = -10$
* **Opposite directions**, hence **negative dot product**.

---

### ‚û°Ô∏è 3. Orthogonal (dot product zero)



````python
import matplotlib.pyplot as plt
import numpy as np

a = np.array([5,0])
b = np.array([0,2])

plt.figure(figsize=(6,6))
plt.quiver(*origin, *a, color='r', scale=1, scale_units='xy', angles='xy', label='a = (5,0)')
plt.quiver(*origin, *b, color='g', scale=1, scale_units='xy', angles='xy', label='b = (0,2)')
plt.grid()
plt.axis('equal')
plt.legend()
plt.title("Dot product zero: vectors at 90 degrees")
plt.show()
````

## üöÄ Application: Cosine similarity in recommendation systems

### üé¨ Example in data science

* Represent movies by feature vectors:

$$
\text{Avengers} = (1,2,0,3,1),\quad \text{Movie B} = (2,0,1,1,1)
$$

* Compute cosine similarity:

$$
\cos \theta = \frac{ \mathbf{a} \cdot \mathbf{b} }{||\mathbf{a}|| \times ||\mathbf{b}||}
$$

* Dot product:

$$
(1*2 + 2*0 + 0*1 + 3*1 + 1*1) = 2 + 0 + 0 + 3 + 1 = 6
$$

* Magnitudes:

$$
||a|| = \sqrt{1^2 + 2^2 + 0^2 + 3^2 + 1^2} = \sqrt{1+4+0+9+1}= \sqrt{15}=3.872
$$

$$
||b|| = \sqrt{2^2 + 0^2 + 1^2 + 1^2 + 1^2} = \sqrt{2^2+0+1+1+1} = \sqrt{7}=2.646
$$

* Cosine similarity:

$$
\cos \theta = \frac{6}{(3.872)(2.646)} = 0.586
$$

‚Üí Movies are **58.6% similar**, can recommend **Movie B** to a user who liked **Avengers**.




## üìù What is element-wise multiplication?

* **Element-wise multiplication** (also called **Hadamard product**) is when you multiply corresponding elements of two vectors to produce a **new vector of the same dimension**.

For vectors

```
A = [1, 2, 3]
B = [3, 4, 5]
```

Element-wise multiplication gives

```
A ‚äô B = [1*3, 2*4, 3*5] = [3, 8, 15]
```

‚úÖ Unlike the **dot product**, this does **not produce a scalar**, it returns a **vector of the same size**.

---

## üí° Where is it used?

* In **feature engineering** for datasets (like calculating discounts).
* In **neural networks**, especially RNNs / LSTMs / GRUs, for gates (forget, input, output gates) which use element-wise multiplication to decide what information to keep or forget.
* In general **EDA (exploratory data analysis)**, creating new features via element-wise multiplication or subtraction.

---

## üöÄ Example from e-commerce

Imagine:

| Product | Cost | Discount (%) |
| ------- | ---- | ------------ |
| A       | 1000 | 0.10         |
| B       | 500  | 0.20         |
| C       | 200  | 0.15         |

* **Element-wise multiplication** gives discount amount:

```
Discount = Cost ‚äô Discount = [1000*0.1, 500*0.2, 200*0.15] = [100, 100, 30]
```

* Then final price is

```
Final = Cost - Discount = [900, 400, 170]
```

# üêç Python example with plots

---

Here‚Äôs a simple **Python script with matplotlib plots** to show this visually.

---

## ‚úèÔ∏è Code


````python


import numpy as np
import matplotlib.pyplot as plt

# Example data
products = ['A', 'B', 'C']
costs = np.array([1000, 500, 200])
discounts = np.array([0.10, 0.20, 0.15])

# Element-wise multiplication to get discount amount
discount_amount = costs * discounts

# Final price after discount
final_price = costs - discount_amount

# -------------------------
# Plotting
# -------------------------
x = np.arange(len(products))

plt.figure(figsize=(10,6))

# Original costs
plt.bar(x - 0.2, costs, width=0.4, label='Original Cost', color='skyblue')

# Discount amounts
plt.bar(x + 0.2, discount_amount, width=0.4, label='Discount Amount', color='lightgreen')

# Final price
for i in range(len(products)):
    plt.text(x[i], final_price[i] + 20, f'Final: {int(final_price[i])}', ha='center', fontsize=10, color='darkblue')

plt.xticks(x, products)
plt.ylabel('Amount ($)')
plt.title('Element-wise multiplication example in Feature Engineering')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

## üé• **Vector Multiplication & Scaling Concepts**

### üöÄ 1. **Scalar multiplication**

* **Definition:** Multiplying a vector by a scalar (single number).

* **Effect:** Each component of the vector is scaled by that scalar.

* **Example:**

  * Given vector $a = (3, 5, 7)$ and scalar $c = 4$,
  * Scalar multiplication gives:

    $$
    c \cdot a = (12, 20, 28)
    $$

* **Geometric intuition:**

  * If a vector is plotted on a coordinate plane, scalar multiplication stretches or shrinks it.
  * E.g., vector $(2, 2)$ multiplied by $2$ becomes $(4, 4)$, doubling its length in the same direction.

---

### üé® 2. **Normalization in images**

* **Problem:** Grayscale image pixels range from $0$ to $255$.

* **Goal:** Normalize pixel values to $0$ to $1$ for easier computation (reducing large numbers).

* **How:**

  $$
  \text{normalized pixel} = \frac{\text{pixel}}{255}
  $$

* This is also an example of **scalar multiplication**.

---

### üìè 3. **Scaling features (standardization example)**

* **Example:** Given a feature vector for height in cm:

  $$
  h = (160, 170, 180)
  $$

* To convert to meters, multiply by $0.01$:

$$
0.01 \times h = (1.6, 1.7, 1.8)
$$

* This is scalar multiplication used to **scale units**, making data ready for machine learning or consistent with other features.

---

### üßÆ 4. **Recap of vector multiplication types**

* **Dot product:** Produces a scalar (inner product).
* **Element-wise multiplication:** Multiply corresponding elements.
* **Scalar multiplication:** Multiply entire vector by a scalar, scaling all components.

## üöÄ **1. Introduction to matrices**

* A **matrix** is a **rectangular array** of numbers (or symbols) arranged in **rows and columns**.
* Usually denoted by a capital letter (like **A**), with elements $a_{ij}$ where:

  * $i$ = row index
  * $j$ = column index

---

## üìä **2. Matrices for data representation**

* In data science, matrices are widely used to represent datasets.
* Example:

  * A dataset with features: **math score, physics score, biology score** for students.
  * Each row = scores for a student (record), each column = a feature.
  * Forms a **3x3 matrix** if you have 3 students and 3 features.

---

## üñº **3. Images as matrices**

* A grayscale image is represented as a matrix of pixel intensities (0 to 255).
* Example:

  * A **3x3 grayscale image** ‚Üí a **3x3 matrix** of pixel values.
  * 0 = black, 255 = white. Values in between are shades of gray.

---

## üßÆ **4. Confusion matrix in machine learning**

* A special **2x2 matrix** used to evaluate classification models.

  * Example:

    * **50** = true positives
    * **10** = false negatives
    * **5** = false positives
    * **35** = true negatives
* Used to compute accuracy, precision, recall.

---

## üß† **5. Matrices in neural networks**

* **Forward propagation** involves **matrix multiplication**:

  * Inputs $x_1, x_2$ multiplied by weights $W$ plus bias $b$.
  * Expressed as:

    $$
    W^T x + b
    $$
* Shapes are important: e.g., 2 inputs connected to 3 neurons ‚Üí **2x3 weight matrix**.

---

## üìà **6. Matrices in linear regression**

* Regression with multiple features (like study hours & IQ predicting exam scores).
* Equation in matrix form:

$$
y = M^T x + c
$$

* $M$ holds coefficients (slopes for each feature), $x$ holds feature values.

## üöÄ **1Ô∏è‚É£ Matrix addition & subtraction**

* ‚úÖ **Definition:** Add or subtract corresponding elements of matrices.
* ‚ö†Ô∏è **Condition:** Matrices must have **same dimensions**.
* **Example:**

  * Matrix A (daily sales in Store A) and Matrix B (Store B) ‚Äî both 3√ó3.
  * Add A + B to get combined sales.

---

## üöÄ **2Ô∏è‚É£ Scalar multiplication**

* ‚úÖ **Definition:** Multiply every element of a matrix by a scalar value.
* **Example use case:**

  * Adjust product prices for **inflation** by multiplying each element by a factor (e.g., 1.05 for 5% inflation).
* Similar technique used in **feature engineering** to scale or adjust data.

---

## üöÄ **3Ô∏è‚É£ Matrix multiplication**

* ‚úÖ **Definition:** Dot product of **rows of first matrix** with **columns of second matrix**.
* ‚ö†Ô∏è **Condition:**

  * If A is of size (m √ó n) and B is (n √ó p), multiplication is possible.
  * Resulting matrix C will be of size (m √ó p).

---

### üîç **Detailed example:**

* Given:

  * Matrix A (2 √ó 3)
  * Matrix B (3 √ó 2) \[after transposing to make multiplication possible]
* Compute each element C<sub>ij</sub> by summing products of row i of A and column j of B.
* Result: C is a **2 √ó 2** matrix.

---

## üí° **Other key points discussed:**

* **Transpose:** Switching rows to columns, needed to align dimensions for multiplication.
* In practice:

  * Used heavily in **deep learning**, **neural networks** (weight & bias operations), **linear regression**, etc.
* All major machine learning algorithms rely on these operations internally.
