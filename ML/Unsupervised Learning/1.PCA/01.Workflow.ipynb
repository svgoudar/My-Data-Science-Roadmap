## üîé **How PCA Works**

PCA (Principal Component Analysis) is a **dimensionality reduction technique** that transforms high-dimensional data into a smaller set of variables (principal components) while preserving as much information (variance) as possible.

---

### 1. **Intuition**

* Imagine you have a cloud of data points in 2D or 3D.
* The data might be stretched more in one direction than others.
* PCA finds those **directions of maximum spread (variance)** and uses them as new axes.

These new axes are called **principal components (PCs)**.

---

### 2. **Steps in PCA**

1. **Standardize the data**

   * Make each feature have mean = 0 and standard deviation = 1.
   * Ensures all features are treated equally, regardless of scale.

2. **Compute the covariance matrix**

   * Describes how features vary together.
   * Example: If height and weight grow together, their covariance is high.

3. **Find eigenvectors and eigenvalues**

   * Perform eigen-decomposition of the covariance matrix.
   * **Eigenvectors ‚Üí directions (principal components).**
   * **Eigenvalues ‚Üí importance (variance explained by each component).**

4. **Rank components**

   * Sort eigenvectors by their eigenvalues (largest to smallest).
   * **PC1** = direction of maximum variance.
   * **PC2** = next most variance, perpendicular to PC1.
   * And so on‚Ä¶

5. **Project the data**

   * Transform original data into this new coordinate system.
   * Keep only the top **k components** that explain most of the variance.

---

### 3. **Key Ideas**

* Each new axis (PC) is **orthogonal (uncorrelated)** to the others.
* PCA doesn‚Äôt ‚Äúremove‚Äù data ‚Äî it **re-expresses** it in fewer dimensions.
* Example: A 100-feature dataset might be reduced to just 10 PCs while keeping 90‚Äì95% of the variance.

---

### 4. **Analogy**

Think of PCA like rotating the camera:

* At first, you see a messy 3D scatter of points.
* By rotating to the right angle (principal components), you can see the data‚Äôs structure more clearly, often in fewer dimensions.

---

‚úÖ **In short:** PCA finds the most informative directions in your data, reorients it along those directions, and lets you drop the less important ones to reduce complexity without losing much information.




### Eigenvectors, Eigenvalues, and Principal Component Analysis (PCA)

The source, *‚ÄúEigenvectors, Eigenvalues, and Principal Component Analysis,‚Äù* explains the mathematical foundations of PCA and highlights the essential role of **eigenvectors** and **eigenvalues** in identifying principal components.

---

#### **1. Role of Eigenvectors and Eigenvalues in PCA**

The central idea is that **eigenvectors and eigenvalues help determine the best principal component line** in PCA.

* The *best* line is the one that **captures the maximum variance** when the dataset is projected onto it.
* It is mathematically proven that the principal component with the greatest variance corresponds directly to the **eigenvector associated with the largest eigenvalue**.

---

#### **2. Linear Transformation and Eigen-decomposition**

To understand this, the source introduces **linear transformations**:

* When a matrix $A$ transforms a vector $v$, the result is another vector that is simply a scaled version of $v$:

$$
Av = \lambda v
$$

Here:

* $v$ = eigenvector (direction that remains unchanged except for scaling)
* $\lambda$ = eigenvalue (the scaling factor)

This relationship forms the basis of **eigen decomposition**, which is applied to the **covariance matrix** in PCA.

* A given matrix can yield multiple eigenvectors, each paired with an eigenvalue.
* The eigenvalue indicates how much variance its corresponding eigenvector captures.

---

#### **3. Linking Eigenvalues to Principal Components**

Among all eigenvectors:

* The one with the **largest eigenvalue** becomes the **first principal component (PC1)** since it captures the maximum variance.
* The next largest eigenvalue corresponds to **PC2**, capturing the next highest variance, and so on.
  Thus, eigenvalues act as a measure of ‚Äúimportance‚Äù for each principal component.

---

#### **4. Practical Steps in PCA**

The source outlines PCA as a sequence of steps:

1. **Standardize the Data**

   * Transform features so they have mean = 0.
   * Prevents features with larger scales from dominating the analysis.

2. **Compute the Covariance Matrix**

   * Captures relationships between features.
   * For two features (x, y), this gives a $2 \times 2$ matrix:

     $$
     \begin{bmatrix}
     Var(x) & Cov(x,y) \\
     Cov(y,x) & Var(y)
     \end{bmatrix}
     $$
   * In general, an $n \times n$ covariance matrix is formed for $n$ features.

3. **Eigen-decomposition of the Covariance Matrix**

   * Solve for eigenvalues $\lambda$ and eigenvectors $v$.
   * Each eigenvector defines a new axis (principal component).
   * Each eigenvalue indicates the variance explained along its axis.

---

#### **5. Dimensionality Reduction with Principal Components**

The eigenvectors and eigenvalues are then used for dimensionality reduction:

* For 2D data ‚Üí two eigenvalues (Œª1, Œª2) are found. Projecting onto the eigenvector with the larger eigenvalue (PC1) reduces data to 1D while preserving maximum variance.
* For 3D data ‚Üí three eigenvalues (Œª1, Œª2, Œª3) are obtained. To reduce to 2D, the top two eigenvectors (PC1, PC2) are chosen; for 1D, only PC1 is used.

Dimensionality reduction is achieved by **projecting the data onto the new principal component axes**.

---

### **Summary**

The source explains that PCA works by:

1. Standardizing data,
2. Computing its covariance matrix, and
3. Performing eigen decomposition.

The resulting eigenvectors (principal components) define new axes, while eigenvalues indicate how much variance each axis captures. By selecting the top components (those with the largest eigenvalues), PCA reduces dimensionality while retaining the maximum possible information.



## üî¢ Example: PCA on a 2D Dataset

Suppose we have a tiny dataset with two features:

$$
X =
\begin{bmatrix}
2 & 0 \\
0 & 2 \\
3 & 1 \\
1 & 3
\end{bmatrix}
$$

---

### **Step 1: Standardize the Data**

* Compute the mean of each column:

  $$
  \mu_x = \frac{2+0+3+1}{4} = 1.5, \quad
  \mu_y = \frac{0+2+1+3}{4} = 1.5
  $$

* Subtract the mean (center the data):

$$
X_{\text{centered}} =
\begin{bmatrix}
0.5 & -1.5 \\
-1.5 & 0.5 \\
1.5 & -0.5 \\
-0.5 & 1.5
\end{bmatrix}
$$

---

### **Step 2: Covariance Matrix**

The covariance matrix is:

$$
C = \frac{1}{n-1} X_{\text{centered}}^T X_{\text{centered}}
$$

$$
C = \frac{1}{3}
\begin{bmatrix}
(0.5)^2 + (-1.5)^2 + (1.5)^2 + (-0.5)^2 & (0.5)(-1.5) + (-1.5)(0.5) + (1.5)(-0.5) + (-0.5)(1.5) \\
(-1.5)(0.5) + (0.5)(-1.5) + (-0.5)(1.5) + (1.5)(-0.5) & (-1.5)^2 + (0.5)^2 + (-0.5)^2 + (1.5)^2
\end{bmatrix}
$$

$$
C = \frac{1}{3}
\begin{bmatrix}
5 & -3 \\
-3 & 5
\end{bmatrix}
$$

$$
C =
\begin{bmatrix}
1.67 & -1.0 \\
-1.0 & 1.67
\end{bmatrix}
$$

---

### **Step 3: Eigen Decomposition**

Solve:

$$
|C - \lambda I| = 0
$$

$$
\begin{vmatrix}
1.67 - \lambda & -1 \\
-1 & 1.67 - \lambda
\end{vmatrix} = 0
$$

$$
(1.67 - \lambda)^2 - 1 = 0
$$

$$
\lambda_1 = 2.67, \quad \lambda_2 = 0.67
$$

* Eigenvalue 1 ($\lambda_1 = 2.67$) explains **most variance**.
* Eigenvalue 2 ($\lambda_2 = 0.67$) explains less variance.

---

### **Step 4: Eigenvectors (Principal Components)**

Solve $Cv = \lambda v$.

For $\lambda_1 = 2.67$:

$$
v_1 =
\begin{bmatrix}
0.71 \\
-0.71
\end{bmatrix}
$$

For $\lambda_2 = 0.67$:

$$
v_2 =
\begin{bmatrix}
0.71 \\
0.71
\end{bmatrix}
$$

---

### **Step 5: Choose Principal Component(s)**

* **PC1** = eigenvector of largest eigenvalue = $[0.71, -0.71]$
* **PC2** = eigenvector of second eigenvalue = $[0.71, 0.71]$

If we reduce from 2D ‚Üí 1D, we **project onto PC1**.

---

### ‚úÖ **Interpretation**

* PC1 captures variance = 2.67 (\~80% of total variance).
* PC2 captures variance = 0.67 (\~20% of total variance).
* So, keeping just PC1 preserves \~80% of the dataset‚Äôs information while reducing dimensionality.



```python
import numpy as np
import matplotlib.pyplot as plt

# Original dataset
X = np.array([
    [2, 0],
    [0, 2],
    [3, 1],
    [1, 3]
])

# Mean-center the data
X_mean = np.mean(X, axis=0)
X_centered = X - X_mean

# Covariance matrix
C = np.cov(X_centered.T)

# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(C)

# Sort eigenvalues and eigenvectors in descending order
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Plot the centered data
plt.figure(figsize=(7,7))
plt.scatter(X_centered[:,0], X_centered[:,1], color='blue', label="Data Points")

# Plot eigenvectors (scaled for visualization)
for i in range(len(eigenvalues)):
    vec = eigenvectors[:,i] * np.sqrt(eigenvalues[i])*2
    plt.plot([0, vec[0]], [0, vec[1]], linewidth=3, label=f'PC{i+1}')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("X1 (centered)")
plt.ylabel("X2 (centered)")
plt.title("PCA Visualization: Data with Principal Components")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
