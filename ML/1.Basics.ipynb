### ðŸŽ“ **Understanding Classification in Supervised Learning**

### ðŸ©º **Example: Breast Cancer Detection**

Letâ€™s understand classification with a practical and impactful example â€” **detecting breast cancer**.

Imagine you're building a machine learning system that helps doctors diagnose breast cancer. Early detection can be life-saving. Your system will use a patient's medical records to determine whether a tumor is:

* **Benign (not dangerous)**
* **Malignant (cancerous)**

* **X-axis**: Tumor size

### ðŸ”¢ **Finite Outputs: The Key to Classification**

Because there are only **two possible outputs**, we call this a **binary classification** problem.

### **Unsupervised Learning Overview**

* **Definition**: In **unsupervised learning**, the algorithm is given only the input data **X**, without corresponding labels **Y**. It must identify patterns or structures in the data on its own.

* **Contrast with Supervised Learning**:

  * Supervised learning uses both **input features (X)** and **output labels (Y)**.
  * Unsupervised learning uses only **input features (X)**.

---

### **Types of Unsupervised Learning**

1. **Clustering**:

   * Groups similar data points together.
   * Example: Grouping news articles on similar topics.

2. **Anomaly Detection**:

   * Identifies unusual or rare events.
   * Common use case: **Fraud detection** in financial systems.

3. **Dimensionality Reduction**:

   * Compresses large datasets into lower-dimensional representations while retaining as much information as possible.
   * Useful for data visualization, compression, and speeding up algorithms.

> Note: If Anomaly Detection or Dimensionality Reduction seems unclear now, it will be explained in depth later in the course.

### **Examples to Distinguish Between Supervised and Unsupervised Learning**

* âœ… **Unsupervised**:

  * Grouping news articles by topic (clustering).
  * Market segmentation using customer data (clustering).

* âŒ **Supervised**:

  * Spam detection (emails labeled as spam or not).
