## ðŸ”¹ **1. Workflow of SVM for Classification (SVC)**

### Step 1: Input Data

* Dataset with features $X$ and categorical labels $y \in \{-1, +1\}$.

### Step 2: Kernel Selection

* Choose kernel (Linear, Polynomial, RBF, Sigmoid) depending on data separability.

### Step 3: Find Decision Boundary

* SVM tries to find a **hyperplane** that separates classes with **maximum margin**.
* Nearest data points to hyperplane = **Support Vectors**.

### Step 4: Optimization Problem

$$
\min_{w,b,\xi} \ \frac{1}{2}||w||^2 + C \sum_{i=1}^n \xi_i
$$

subject to

$$
y_i (w^T x_i + b) \geq 1 - \xi_i, \quad \xi_i \geq 0
$$

* $C$: Regularization parameter (controls tradeoff between margin and misclassification).
* $\xi_i$: Slack variable (allows misclassification for soft margin).

### Step 5: Prediction

* Classify new data using:

$$
\hat{y} = \text{sign}(w^T x + b)
$$

## ðŸ”¹ **2. Workflow of SVM for Regression (SVR)**

### Step 1: Input Data

* Dataset with features $X$ and continuous labels $y \in \mathbb{R}$.

### Step 2: Kernel Selection

* Choose kernel (Linear, RBF, Polynomial) depending on data distribution.

### Step 3: Define Îµ-insensitive Tube

* Instead of a separating hyperplane, SVR fits a function $f(x) = w^T x + b$
* Aims to keep predictions within an **Îµ-margin (tube)** around actual values.

### Step 4: Optimization Problem

$$
\min_{w,b,\xi, \xi^*} \ \frac{1}{2}||w||^2 + C \sum_{i=1}^n (\xi_i + \xi_i^*)
$$

subject to:

$$
y_i - (w^T x_i + b) \leq \epsilon + \xi_i
$$

$$
(w^T x_i + b) - y_i \leq \epsilon + \xi_i^*, \quad \xi_i, \xi_i^* \geq 0
$$

* $Îµ$: Tolerance zone (error within Îµ is ignored).
* $\xi_i, \xi_i^*$: Slack variables for deviations outside the tube.

### Step 5: Prediction

* Predict values using regression function:

$$
\hat{y} = w^T x + b
$$
