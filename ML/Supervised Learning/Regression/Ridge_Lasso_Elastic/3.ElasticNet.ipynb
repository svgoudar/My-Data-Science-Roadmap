## üîπ What is Elastic Net Regression?

**Elastic Net Regression** is a **regularized linear regression** technique that combines the penalties of:

* **Lasso (L1)** ‚Üí drives some coefficients to **zero** (feature selection).
* **Ridge (L2)** ‚Üí shrinks coefficients (handles multicollinearity).

It is especially useful when you have **many correlated features**.

---

## üîπ Elastic Net Loss Function

For a regression model:

$$
y = X\beta + \epsilon
$$

The Elastic Net objective function is:

$$
L(\beta) = \frac{1}{2n} \sum_{i=1}^n (y_i - \hat{y}_i)^2 + \lambda \left( \alpha \sum_{j=1}^p |\beta_j| + (1-\alpha)\sum_{j=1}^p \beta_j^2 \right)
$$

Where:

* $y_i$ ‚Üí actual value
* $\hat{y}_i$ ‚Üí predicted value
* $\beta_j$ ‚Üí regression coefficients
* $\lambda$ ‚Üí overall regularization strength
* $\alpha$ ‚Üí mixing parameter between L1 and L2

  * If $\alpha = 1$: becomes **Lasso**
  * If $\alpha = 0$: becomes **Ridge**

---

## üîπ Why Use Elastic Net?

* **Lasso issue**: If features are highly correlated, it tends to pick one and ignore others ‚Üí unstable.
* **Ridge issue**: Keeps all features but doesn‚Äôt perform feature selection.
* **Elastic Net**: Combines both strengths ‚Üí
  ‚úÖ Keeps model stable with correlated features.
  ‚úÖ Performs feature selection.

---

## üîπ Example Intuition

Suppose you‚Äôre predicting **house price** using:

* `square_feet`
* `bedrooms`
* `bathrooms`
* `location_score`

Since `square_feet` and `bedrooms` are highly correlated:

* **Lasso** may drop `bedrooms` entirely.
* **Ridge** will keep both but shrink their weights.
* **Elastic Net** ‚Üí keeps both but controls their weights ‚Üí better balance.

## üîπ Pros & Cons

‚úÖ Handles multicollinearity
‚úÖ Performs feature selection
‚úÖ Works well when $p > n$ (more features than samples)
‚ùå Needs tuning of two hyperparameters ($\lambda, \alpha$)
