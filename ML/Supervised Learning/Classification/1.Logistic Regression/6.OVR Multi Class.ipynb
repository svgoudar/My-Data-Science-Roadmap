### ðŸ”¹ One-vs-Rest (OvR) Approach

1. **Idea**: Train multiple binary classifiers, one for each class.

   * For class 1 vs (class 2 + class 3).
   * For class 2 vs (class 1 + class 3).
   * For class 3 vs (class 1 + class 2).
2. Each model outputs a probability that a new sample belongs to its target class.
3. The **final prediction** = class with the highest probability.
