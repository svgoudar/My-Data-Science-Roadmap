The transcript explains why **linear regression is not suitable for binary classification** and motivates the need for **logistic regression**.

### Key Points:

1. **Binary Classification Definition**

   * Output variable has two categories (e.g., pass/fail).
   * Example: Study hours → predict pass (1) or fail (0).

2. **Using Linear Regression for Classification**

   * One can fit a best-fit line and apply a threshold (≤0.5 → 0, >0.5 → 1).
   * This seems to work initially.

3. **Problems with Linear Regression in Classification**

   * **Outliers**: A single outlier can shift the best-fit line, changing predictions drastically.
   * **Invalid Outputs**: Predictions can be less than 0 or greater than 1, which are not valid probabilities for classification.

4. **Need for Logistic Regression**

   * Logistic regression applies a **sigmoid (squashing) function** that restricts outputs between 0 and 1.
   * This makes predictions interpretable as probabilities and robust against the problems faced with linear regression.

**Conclusion**: Linear regression fails for classification due to sensitivity to outliers and unbounded outputs. Logistic regression solves this by mapping predictions to the \[0,1] range using the sigmoid function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions
x = np.linspace(-5, 5, 400)
convex_func = x**2
nonconvex_func = np.sin(x)

# Plot convex function
plt.figure(figsize=(10,4))
plt.plot(x, convex_func, label=r"$f(x) = x^2$ (Convex)")
plt.title("Convex Function Example")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

# Plot non-convex function
plt.figure(figsize=(10,4))
plt.plot(x, nonconvex_func, label=r"$f(x) = \sin(x)$ (Non-Convex)", color="red")
plt.title("Non-Convex Function Example")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
