Great! Let's dive into **`ewm()` — Exponentially Weighted Moving** calculations in Pandas.

---

### ✅ What is `ewm()`?

`ewm()` stands for **Exponentially Weighted Moving** functions. It is a method used to calculate moving averages or statistics that give **more weight to recent observations** while not completely ignoring older ones.

This is **very useful in time-series** or streaming data — like tracking recent trends, stock prices, server metrics, etc.

---

### 📌 Why use EWM instead of Rolling?

* `.rolling(window=3).mean()` → averages equally across the last 3 values.
* `.ewm(span=3).mean()` → gives **exponentially decreasing weights**, with **most weight on the most recent** value.

---

### 🧠 Core Formula (Simplified):

For a series of values $x_1, x_2, ..., x_t$, the **exponential moving average** $EMA_t$ is calculated recursively:

$$
EMA_t = \alpha \cdot x_t + (1 - \alpha) \cdot EMA_{t-1}
$$

Where:

* $\alpha$ is the **smoothing factor**: $\alpha = \frac{2}{\text{span} + 1}$

---

### 🔧 Syntax in Pandas:

```python
df['ewm_avg'] = df['your_column'].ewm(span=3).mean()
```

---

### ⚙️ Parameters You Should Know:

| Parameter     | Meaning                                             |
| ------------- | --------------------------------------------------- |
| `span`        | Defines the decay; higher span = slower decay       |
| `alpha`       | Directly specify smoothing factor (instead of span) |
| `adjust=True` | Whether to adjust weights (like in stats)           |
| `ignore_na`   | Whether to ignore missing values in the calculation |

---

### 📊 Example:

```python
import pandas as pd

data = pd.Series([10, 20, 30, 40, 50])
ema = data.ewm(span=3, adjust=False).mean()

print(ema)
```

**Output:**

```
0    10.000000
1    15.000000
2    22.500000
3    31.250000
4    40.625000
dtype: float64
```

You can see how each value is **closer to the most recent ones**.

---

### 🔍 When to Use EWM?

* Financial indicators (e.g., EMA for stock prices)
* Real-time sensor or monitoring data
* Smoothing noisy time series
* Trend detection in recent data

---

## 🔹 8. **Reshaping / Pivoting**

### ✅ 1. `.pivot()`

Used to reshape data (similar to Excel pivot). Requires **unique index/column pairs**.

````python
df_uni = df.drop_duplicates(subset='customer_id')
````

````python
from IPython.display import Markdown, display
display(Markdown("**Effect**: Shows loan amounts requested by each customer for different loan types (1 row per customer)."))
df_uni.pivot(index='customer_id', columns='loan_type', values='loan_amount_requested')
````

### ✅ 2. `.pivot_table()`
Like `.pivot()`, but allows **aggregation** for duplicate entries using a function like `mean()`.


````python
from IPython.display import Markdown, display
display(Markdown("**Effect**: Average loan amount requested per employment type for each loan type.\n"))
df_pivot = df_uni.pivot_table(index='employment_status', columns='loan_type', values='loan_amount_requested', aggfunc='mean')
df_pivot
````

### ✅ 3. `.melt()`
Unpivots a wide table to long format. Useful for column-to-row transformation.



````python
df_pivot.reset_index(inplace=True)
````

````python
df_long = pd.melt(
    df_pivot,
    id_vars='employment_status',
    var_name='loan_type',
    value_name='avg_loan_amount_requested'
)
df_long
````
