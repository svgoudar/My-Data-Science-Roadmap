| API                    | What it does                    |
| ---------------------- | ------------------------------- |
| `head() / tail()`      | See top / bottom rows           |

| `isnull(), sum()`      | Missing value analysis          |
| `unique(), nunique()`  | Unique value exploration        |
| `value_counts()`       | Counts of each category         |
| `memory_usage()`       | Memory footprint                |
| `sample()`             | Random sampling of data         |
