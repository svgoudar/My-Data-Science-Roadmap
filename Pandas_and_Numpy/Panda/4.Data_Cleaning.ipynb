| Task                       | Example usage                               |
| -------------------------- | ------------------------------------------- |
| Detect missing             | `df.isnull()`, `df.isnull().sum()`          |
| Remove rows/cols with NaNs | `df.dropna()`, `df.dropna(axis=1)`          |
| Fill missing values        | `df.fillna(0)`, `df.fillna(method='ffill')` |
| Replace specific values    | `df.replace('?', np.nan)`                   |
| Drop duplicate rows        | `df.drop_duplicates()`                      |
| Change data type           | `df['col'].astype(int)`                     |
| Clean strings              | `df['col'].str.strip().str.lower()`         |


| API                     | Description                                                          | Example                        |
| ----------------------- | -------------------------------------------------------------------- | ------------------------------ |
| `isnull()` / `isna()`   | Detect missing values → boolean DataFrame. Same function, two names. | `df.isnull()` or `df.isna()`   |
| `notnull()` / `notna()` | Detect non-missing values.                                           | `df.notnull()` or `df.notna()`
| `dropna()` | Drop rows or columns with NaNs. | `df.dropna()` drops rows with any NaNs.                    |
|            |                                 | `df.dropna(axis=1)` drops columns.                         |
|            |                                 | `df.dropna(how='all')` drops only if all NaN.              |
|            |                                 | `df.dropna(thresh=3)` keeps rows with at least 3 non-NaNs. |
| `fillna()` | Fill NaNs with a scalar, dict, method. | `df.fillna(0)` replaces NaNs with 0.            |
|            |                                        | `df.fillna({'col1': 0, 'col2':99})` per column. |
|            |                                        | `df.fillna(method='ffill')` forward fill.       |
|            |                                        | `df.fillna(method='bfill')` backward fill.      |
| `replace()` | Replace any specified value. | `df.replace(-1, np.nan)` converts `-1` to `NaN`. |
|             |                              | `df.replace(['?', 'NA'], np.nan)`                |
| `interpolate()` | Fill NaNs by interpolation. | `df['col'].interpolate()` does linear interpolation. |
| `dropna(subset=...)` | Drop rows where specific cols are NaN.             | `df.dropna(subset=['col1'])` |
| `where`              | Keep values where condition is True; else set NaN. | `df.where(df > 0)`           |
| `mask`               | Opposite of `where`: set NaN where condition True. | `df.mask(df < 0)`            |
