```python
import pandas as pd
df_loan= pd.read_csv("https://raw.githubusercontent.com/svgoudar/datasets/refs/heads/main/loan_applications.csv",)
df_tran = pd.read_csv("https://raw.githubusercontent.com/svgoudar/datasets/refs/heads/main/transactions.csv")
```

```python
df  = pd.merge(df_loan,df_tran,on='customer_id')
```

1. `.map()` – elementwise mapping on Series(for panda < 2.1.0)

for < 2.1.0 Apply a function to a Dataframe elementwise.



```python
df['gender_mapped'] = df['gender'].map({'Male': 0, 'Female': 1})
```

```python
df.map(lambda x : len(str(x)), na_action='ignore')
```

```python
df_1 = df[['debt_to_income_ratio','interest_rate_offered']]
```

```python
df_1.map(round,ndigits=1)
```

2. `.apply()` – row/column-wise operations


```python
# Calculate a custom risk score
df['risk_score'] = df.apply(
    lambda row: row['loan_amount_requested'] / (row['monthly_income'] + 1),
    axis=1
)
```
