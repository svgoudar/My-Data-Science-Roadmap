```python
import pandas as pd
df_loan= pd.read_csv("https://raw.githubusercontent.com/svgoudar/datasets/refs/heads/main/loan_applications.csv",)
df_tran = pd.read_csv("https://raw.githubusercontent.com/svgoudar/datasets/refs/heads/main/transactions.csv")
```

```python
df  = pd.merge(df_loan,df_tran,on='customer_id')
```

2. `.apply()` – row/column-wise operations


```python
# Calculate a custom risk score
df['risk_score'] = df.apply(
    lambda row: row['loan_amount_requested'] / (row['monthly_income'] + 1),
    axis=1
)
```

3. `.pipe()` – function chaining

```python
def add_cibil_grade(df):
    df['cibil_grade'] = df['cibil_score'].apply(lambda x: 'Excellent' if x > 750 else 'Good' if x > 650 else 'Poor')
    return df

# Use pipe
df = df.pipe(add_cibil_grade)
```

```python
df
```

5. `.eval()` – efficient expressions

```python
df['existing_emis_monthly'] = df['existing_emis_monthly'].astype(int)
df['monthly_income'] = df['monthly_income'].astype(int)
```

```python
df.head(1)
```

```python
df.eval("emi_burden_ratio = existing_emis_monthly / (monthly_income + 1)", inplace=True)
```

6. `.assign()` – add new columns in chain

```python
# Add new columns while preserving chaining
df = df.assign(
    is_high_risk=lambda x: x['risk_score'] > 15,
    age_bracket=lambda x: pd.cut(x['applicant_age'], bins=[0, 25, 35, 50, 100], labels=['Young', 'Mid-Age', 'Adult', 'Senior'])
)
```

```python
df.head(1)
```

7. `.transform()` – preserve shape with transformations

```python
# Normalize monthly income (min-max scaling)
df['income_scaled'] = df[['monthly_income']].transform(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
df['loan_amount_zscore'] = df.groupby('loan_type')['loan_amount_requested'].transform(
    lambda x: (x - x.mean()) / x.std()
)
```

| Feature            | `transform()`                            | `apply()`                            |
| ------------------ | ---------------------------------------- | ------------------------------------ |
| Output shape       | Same as input (can assign to new column) | Can shrink or expand                 |
| Common use case    | Feature engineering                      | Group-wise aggregations or reshaping |
| Fast and optimized | ✅                                        | ❌ Usually slower                     |
