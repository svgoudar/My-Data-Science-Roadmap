| Feature          | `.loc`                        | `.iloc`                      |
| ---------------- | ----------------------------- | ---------------------------- |
| Indexes by       | **Label names**               | **Integer positions**        |
| Works with       | Index names & col names       | Row/col numbers (0,1,2)      |
| Slices are       | **Inclusive** of end          | **Exclusive** of end         |
| Typical use case | When index/columns have names | When treating as pure matrix |

| Function   | Used for                       | Example                                        |
| ---------- | ------------------------------ | ---------------------------------------------- |
| `query()`  | Row filter by string expr      | `df.query('Age > 30 and Gender=="Male"')`      |
| `filter()` | Select cols/rows by name/regex | `df.filter(regex='^loan', axis=1)`             |
| `isin()`   | Check membership               | `df[df['loan_type'].isin(['Car', 'Edu'])]`     |
| `at[]`     | Single value by label          | `df.at['CUST109427', 'loan_amount_requested']` |


| Purpose                 | Example                            |
| ----------------------- | ---------------------------------- |
| Single col select       | `df['Name']`                       |
| Multi col select        | `df[['Name', 'Age']]`              |
| Row by label            | `df.loc[2]`                        |
| Row by position         | `df.iloc[2]`                       |
| Row+col slice by label  | `df.loc[2, 'Name']`                |
| Row+col slice by pos    | `df.iloc[2, 0]`                    |
| Boolean condition       | `df[df['Age'] > 30]`               |
| Query string            | `df.query('Age > 30')`             |
| Filter columns by regex | `df.filter(regex='^A', axis=1)`    |
| Membership check        | `df[df['Name'].isin(['A','B'])]`   |
| Fast scalar access      | `df.at[2, 'Name']` / `df.iat[2,0]` |
